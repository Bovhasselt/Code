```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
train = pd.read_csv('Data/train.csv')
test = pd.read_csv('Data/test.csv')
display(train.head())
```

```python
train['Time'] = train['t']
train.drop(columns=['t'], inplace=True)
display(train.head())

test['Time'] = test['t']
test.drop(columns=['t'], inplace=True)
display(test.head())
```

```python
# add these two dataframes together
df = pd.concat([train, test], axis=0)

# set time as datetime starting from 2023-05-08 10:54:00 and increasing by 1 minute every time
df['Time'] = pd.date_range(start='2023-05-08 10:54:00', periods=len(df), freq='1min')
df.drop(columns=['Water-1', 'Water-2', 'Water-3'], inplace=True)
display(df.head())
```

```python
fig = px.line(df, x='Time', y=['Water', 'Hot gas pipe', 'Liquid pipe', 'Suction pipe',
       'Door', 'Fridge'], title='Time Series with Rangeslider Small Vat')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
df.to_csv('Data/plot_data_smallvat.csv', index=False)
```
